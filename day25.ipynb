```python
import os
import numpy as np

def read_file(filename='input.txt', path=os.path.join(os.getcwd())):
    with open(os.path.join(path, filename), "r") as f:
        input_list = [line.strip() for line in f]
    return input_list
```

# Day 25

## Part 1

```python
input25 = read_file('input25.txt')
data = np.array([list(l) for l in input25])

cucumbers_moved = True
n, m = data.shape

step_count = 0
while cucumbers_moved:
    cucumbers_moved = False
    
    new_data = np.copy(data)
    for i in range(n):
        for j in range(m):
            if data[i, j] == '>' and data[i, (j+1)%m] == '.':
                cucumbers_moved = True
                #Move cucumber
                new_data[i, (j+1)%m], new_data[i, j] = new_data[i, j], new_data[i, (j+1)%m]
    
    #Update data for southern cucumbers
    data = new_data
    new_data = np.copy(data)

    for j in range(m):
        for i in range(n):
            if data[i, j] == 'v' and data[(i+1)%n, j] == '.':
                cucumbers_moved = True
                #Move cucumber
                new_data[(i+1)%n, j], new_data[i, j] = new_data[i, j], new_data[(i+1)%n, j]

    data = new_data
    step_count += 1
    
print(step_count)
```

```python
import os

def read_file(filename='input.txt', path=os.path.join(os.getcwd())):
    with open(os.path.join(path, filename), "r") as f:
        input_list = [line.strip() for line in f]
    return input_list
```

## Part 1

```python
def find_explode(sn):
    
    sn_str = str(sn)
    
    depth = 0
    
    for i in range(len(sn_str)):
        if sn_str[i] == '[':
            depth += 1
        elif sn_str[i] == ']':
            depth -= 1
        
        if depth == 5:
            return i
    
    return None

def explode(sn, i):
    
    exploded_sn_str = ''
    sn_str = str(sn)
    
    l = i
    parenthesis_not_found = True
    while parenthesis_not_found:
        if sn_str[l] == ']':
            parenthesis_not_found = False
        else:
            l += 1
        
    pair = eval(sn_str[i:l+1])
    
    left_side = sn_str[:i]
    right_side = sn_str[l+1:]
    
    j = len(left_side)-1
    no_left_num = True
    while j >= 0 and no_left_num:
        if left_side[j].isnumeric():
            l = j-1
            is_numeric = True
            while is_numeric:
                if left_side[l].isnumeric():
                    l-=1
                else:
                    is_numeric = False
            left_num = int(left_side[l+1:j+1])
            left_side = left_side[:l+1] + str(left_num+pair[0]) + left_side[j+1:]
            no_left_num = False
        else:
            j -= 1
    
    k = 0
    no_right_num = True
    while k < len(right_side) and no_right_num:
        if right_side[k].isnumeric():
            l = k+1
            is_numeric = True
            while is_numeric:
                if right_side[l].isnumeric():
                    l+=1
                else:
                    is_numeric = False
            right_num = int(right_side[k:l])
            right_side = right_side[:k] + str(right_num+pair[1]) + right_side[l:]
            no_right_num = False
        else:
            k += 1

    return eval(left_side + '0' + right_side) 
    
def find_split(sn):
    
    sn_str = str(sn)
    
    for i in range(len(sn_str)):
        if sn_str[i].isnumeric():
            j = i+1
            while sn_str[j].isnumeric():
                j += 1
            num = int(sn_str[i: j])
            if num >= 10:
                new_pair = [num//2, num-num//2]
                sn_str = sn_str[:i] + str(new_pair) + sn_str[j:]
                return sn_str, True
    return eval(sn_str), False
    
def reduce(sn):
    
    do_explode = True
    do_split = True
    
    while do_explode or do_split:
        i = find_explode(sn)
        if i:
            sn = explode(sn, i)
        else:
            do_explode = False
            sn, do_split = find_split(sn)
            
    return sn

def compute_magnitude(sn):
    
    if type(sn) == int:
        return sn
    
    return 3*compute_magnitude(sn[0])+2*compute_magnitude(sn[1])
```

```python
input18 = read_file('input18.txt')

data = [eval(line) for line in input18]

sn = data[0]
for i in range(1, len(data)):
    sn = reduce([sn, data[i]])

print(f"Magnitude of result is {compute_magnitude(sn)}")
```

## Part 2

```python
input18 = read_file('input18.txt')

data = [eval(line) for line in input18]

m = 0
for i in range(len(data)):
    sn1 = data[i]
    for j in range(len(data)):
        if i != j:
            sn2 = data[j]
            sn = reduce([sn1, sn2])
            magnitude = compute_magnitude(sn)
            m = max(magnitude, m)
print(f"Maximum magnitude adding two numbers is {m}")
```

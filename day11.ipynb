```python
import os

def read_file(filename='input.txt', path=os.path.join(os.getcwd())):
    with open(os.path.join(path, filename), "r") as f:
        input_list = [line.strip() for line in f]
    return input_list
```

# Day 11

## Part 1

```python
input11 = read_file('input11.txt')
input11 = ['5483143223',
'2745854711',
'5264556173',
'6141336146',
'6357385478',
'4167524645',
'2176841721',
'6882881134',
'4846848554',
'5283751526']


data = []
for row in input11:
    data.append([int(octopus) for octopus in row])

steps = 100

rows = len(data)
cols = len(data[0])

flashes = 0
for step in range(steps):
    #Increase by one all octopuses
    data = [[octopus + 1 for octopus in row] for row in data]
    #Start flashes
    flashed_pos = []
    must_flash = True
    while must_flash:
        for i in range(rows):
            for j in range(cols):
                #If element has to flash
                if data[i][j] > 9:
                    flashes += 1
                    flashed_pos.append((i, j))
                    data[i][j] = 0
                    #If has top element
                    if i-1 >= 0 and (i-1, j) not in flashed_pos:
                        data[i-1][j] += 1
                    #If has top diagonal left
                    if i-1 >= 0 and j-1 >= 0 and (i-1, j-1) not in flashed_pos:
                        data[i-1][j-1] += 1
                    #If has top diagonal right
                    if i-1 >= 0 and j+1 < cols and (i-1, j+1) not in flashed_pos:
                        data[i-1][j+1] += 1
                    #If has left
                    if j-1 >= 0 and (i, j-1) not in flashed_pos:
                        data[i][j-1] += 1
                    #If has right
                    if j+1 < cols and (i, j+1) not in flashed_pos:
                        data[i][j+1] += 1
                    #If has bottom
                    if i+1 < rows and (i+1, j) not in flashed_pos:
                        data[i+1][j] += 1
                    #If has bottom diagonal left
                    if i+1 < rows and j-1 >= 0 and (i+1, j-1) not in flashed_pos:
                        data[i+1][j-1] += 1
                    #If has bottom diagonal right
                    if i+1 < rows and j+1 < cols and (i+1, j+1) not in flashed_pos:
                        data[i+1][j+1] += 1
        data_flash = [[octopus > 9 for octopus in row] for row in data]
        must_flash = True in [True in row for row in data_flash]

print(f"Total flashes after {steps} steps is {flashes}")
```

```python
import os

def read_file(filename='input.txt', path=os.path.join(os.getcwd())):
    with open(os.path.join(path, filename), "r") as f:
        input_list = [line.strip() for line in f]
    return input_list
```

# Day 8

## Part 1

```python
input8 = read_file('input8.txt')
outputs = [out.split(' | ')[1] for out in input8]
outs = [out.split(' ') for out in outputs]

lengths_to_count = [2, 4, 3, 7]
count = 0
for out in outs:
    for o in out:
        if len(o) in lengths_to_count:
            count += 1

print(count)
```

## Part 2

```python
def find_output_number(line_inputs, line_outputs):
    digits = {}
    
    one = ''
    four = ''
    eight = ''
    seven = ''
    
    for input_digit in line_inputs:
        if len(input_digit) == 2:
            one = input_digit
        elif len(input_digit) == 4:
            four = input_digit
        elif len(input_digit) == 7:
            eight = input_digit
        elif len(input_digit) == 3:
            seven = input_digit
    
    number = ''
    for output in line_outputs:
        if len(output) == 2:
            number += '1'
        elif len(output) == 4:
            number += '4'
        elif len(output) == 7:
            number += '8'
        elif len(output) == 3:
            number += '7'
        elif len(output) == 6:
            # si cuatro incluido = 9
            is_nine = 0 not in [c in output for c in four]
            if is_nine:
                number += '9'
            else:
                # si cuatro sin el uno = 6
                four_no_one = four.translate({ord(i): None for i in one})
                is_zero = 0 not in [c in output for c in four_no_one]
                if is_zero:
                    number += '6'
                # 0
                else:
                    number += '0'
        elif len(output) == 5:
            # si uno = 3
            is_three = 0 not in [c in output for c in one]
            if is_three:
                number += '3'
            else:
                # si cuatro sin el uno = 5
                four_no_one = four.translate({ord(i): None for i in one})
                is_five = 0 not in [c in output for c in four_no_one]
                if is_five:
                    number += '5'
                # 2
                else:
                    number += '2'
            
    
    return int(number)
```

```python
input8 = read_file('input8.txt')
data = []
for d in input8:
    inp, out = d.split(' | ')
    data.append((inp.split(' '), out.split(' ')))

suma = 0

for line in data:
    suma += find_output_number(line[0], line[1])

print(suma)
```

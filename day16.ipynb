```python
import os
import numpy as np

def read_file(filename='input.txt', path=os.path.join(os.getcwd())):
    with open(os.path.join(path, filename), "r") as f:
        input_list = [line.strip() for line in f]
    return input_list
```

## Part 1

```python
def hex_to_bin(data):
    data_bin = ""
    for c in data:
        data_bin += bin(int(c, 16))[2:].zfill(4)
    return data_bin
```

```python
def read_literal_packet(data, pos):
    next_packet = True
    packet_bin = ''
    while next_packet:
        next_packet = data[pos] == '1'
        packet_bin += data[pos+1:pos+5]
        pos += 5
    
    return pos, int(packet_bin, 2)

def read_operator_packet(data, pos):
    length_type_id = int(data[pos], 2)
    pos += 1
    sum_versions = 0
    
    if length_type_id == 0:
        length_in_bits = int(data[pos:pos+15], 2)
        pos += 15
        final_pos = pos + length_in_bits
        while pos < final_pos and pos < len(data):
            pos, version = read_packet(data, pos)
            sum_versions += version
    else:
        n_subpackets = int(data[pos:pos+11], 2)
        pos += 11
        for _ in range(n_subpackets):
            pos, version = read_packet(data, pos)
            sum_versions += version
            
    return pos, sum_versions

def read_packet(data, pos):
    version = int(data[pos:pos+3], 2)
    pos += 3
    type_id = int(data[pos:pos+3], 2)
    pos += 3
    sum_version = 0
    if type_id == 4:
        pos, packet = read_literal_packet(data, pos)
    else:
        pos, sum_version = read_operator_packet(data, pos)
        
    return pos, version+sum_version
```

```python
data = read_file('input16.txt')[0]
data_bin = hex_to_bin(data)
pos = 0

pos, sum_versions = read_packet(data_bin, pos)

print(f"The sum of the versions of all the packets is {sum_versions}.")
```

## Part 2

```python
def read_literal_packet(data, pos):
    next_packet = True
    packet_bin = ''
    while next_packet:
        next_packet = data[pos] == '1'
        packet_bin += data[pos+1:pos+5]
        pos += 5
    
    return pos, int(packet_bin, 2)

def read_operator_packet(data, pos, type_id):
    length_type_id = int(data[pos], 2)
    pos += 1
    
    packets = []
    
    if length_type_id == 0:
        length_in_bits = int(data[pos:pos+15], 2)
        pos += 15
        final_pos = pos + length_in_bits
        while pos < final_pos and pos < len(data):
            pos, packet = read_packet(data, pos)
            packets.append(packet)
    else:
        n_subpackets = int(data[pos:pos+11], 2)
        pos += 11
        for _ in range(n_subpackets):
            pos, packet = read_packet(data, pos)
            packets.append(packet)
    
    if type_id == 0:
        packet = sum(packets)
        
    elif type_id == 1:
        packet = 1
        for p in packets:
            packet *= p
            
    elif type_id == 2:
        packet = min(packets)
        
    elif type_id == 3:
        packet = max(packets)
        
    elif type_id == 5:
        packet = packets[0] > packets[1]
    
    elif type_id == 6:
        packet = packets[0] < packets[1]
        
    elif type_id == 7:
        packet = packets[0] == packets[1]
    
            
    return pos, packet

def read_packet(data, pos):
    version = int(data[pos:pos+3], 2)
    pos += 3
    type_id = int(data[pos:pos+3], 2)
    pos += 3
    if type_id == 4:
        pos, packet = read_literal_packet(data, pos)
    else:
        pos, packet = read_operator_packet(data, pos, type_id)
        
    return pos, packet
```

```python
data = read_file('input16.txt')[0]
data_bin = hex_to_bin(data)

pos, packet = read_packet(data_bin, 0)


print(f"Packet is: {packet}.")
```

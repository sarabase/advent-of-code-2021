```python
import os
import numpy as np

def read_file(filename='input.txt', path=os.path.join(os.getcwd())):
    with open(os.path.join(path, filename), "r") as f:
        input_list = [line.strip() for line in f]
    return input_list
```

# Day 22

## Part 1

```python
def preprocess_input(input_data):
    instruct = []
    for line in input_data:
        ins, coord = line.split(' ')
        x, y, z = coord.split(',')
        x = x.split('=')[1]
        y = y.split('=')[1]
        z = z.split('=')[1]
        instruct.append((ins, (int(x.split('..')[0]), int(x.split('..')[1])), (int(y.split('..')[0]), int(y.split('..')[1])), (int(z.split('..')[0]), int(z.split('..')[1]))))
    return instruct

def in_region(ins, region):
    x = ins[1]
    y = ins[2]
    z = ins[3]
    
    if x[0] < region[0][0] or x[1] > region[0][1]:
        return False
    if y[0] < region[1][0] or y[1] > region[1][1]:
        return False
    if z[0] < region[2][0] or z[1] > region[2][1]:
        return False
    return True
```

```python
input22 = read_file('input22.txt')
instruct = preprocess_input(input22)
grid = np.zeros((100, 100, 100))

#0 is off, 1 is on
region = ((-50, 50), (-50, 50), (-50, 50))
for ins in instruct:
    if in_region(ins, region):
        for x in range(ins[1][0], ins[1][1]+1):
            for y in range(ins[2][0], ins[2][1]+1):
                for z in range(ins[3][0], ins[3][1]+1):
                    if ins[0] == 'on':
                        grid[x+50, y+50, z+50] = 1
                    else:
                        grid[x+50, y+50, z+50] = 0
print(f'Considering cubes in the region x=-50..50,y=-50..50,z=-50..50, after going throught the reboot steps, {np.count_nonzero(grid == 1)} cubes are on.')
```

## Part 2

```python
import math
def find_grid(instruct):
    max_x = 0
    min_x = 0 
    max_y = 0 
    min_y = 0 
    max_z = 0
    min_z = 0
    for ins in instruct:
        if ins[1][0] < min_x:
            min_x = ins[1][0]
        if ins[1][1] > max_x:
            max_x = ins[1][1]
        
        if ins[2][0] < min_y:
            min_y = ins[2][0]
        if ins[2][1] > max_y:
            max_y = ins[2][1]
            
        if ins[3][0] < min_z:
            min_z = ins[3][0]
        if ins[3][1] > max_z:
            max_z = ins[3][1]
            
    grid = np.zeros((abs(min_x)+max_x, abs(min_y)+max_y, abs(min_z)+max_z), dtype=int)
    diff = (abs(min_x), abs(min_y), abs(min_z))
    return grid, diff
```

```python
input22 = read_file('input22.txt')
instruct = preprocess_input(input22)
grid, diff = find_grid(instruct)
grid = np.zeros((100, 100, 100))

#0 is off, 1 is on
for ins in instruct:
    for x in range(ins[1][0], ins[1][1]+1):
        for y in range(ins[2][0], ins[2][1]+1):
            for z in range(ins[3][0], ins[3][1]+1):
                if ins[0] == 'on':
                    grid[x+diff[0], y+diff[1], z+diff[2]] = 1
                else:
                    grid[x+diff[0], y+diff[1], z+diff[2]] = 0
print(f'Considering cubes in the region x=-50..50,y=-50..50,z=-50..50, after going throught the reboot steps, {np.count_nonzero(grid == 1)} cubes are on.')
```

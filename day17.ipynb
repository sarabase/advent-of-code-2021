```python
import os

def read_file(filename='input.txt', path=os.path.join(os.getcwd())):
    with open(os.path.join(path, filename), "r") as f:
        input_list = [line.strip() for line in f]
    return input_list
```

```python
def perform_shoot(x, y, min_x, max_x, min_y, max_y):
    highest_y = min_y
    probe_x = 0
    probe_y = 0
    
    target_hit = False
    
    while probe_x < max_x and probe_y > min_y and not target_hit:
        #Update position
        probe_x += x
        probe_y += y
        #Update_velocity
        if x > 0:
            x -= 1
        elif x < 0:
            x += 1
        y -= 1
        target_hit = min_x <= probe_x <= max_x and min_y <= probe_y <= max_y
        highest_y = probe_y if probe_y > highest_y else highest_y
        
    if target_hit:
        return highest_y
    else:
        return False
        
```
